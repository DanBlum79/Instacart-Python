```python
# 4. Transform and loc
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
ords_prods_merge.head(15)
```

```python
ords_prods_merge.head(100)
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(10)
```

```python
# loc
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# 5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer).
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

New customers pay the most in average prices, followed by regular customers, and then loyal customers. Each pay the same as far as the minimum and maximum prices of products.

```python
# 6. Create a spending flag for each user based on the average price across all their orders.
# create new column
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check output
ords_prods_merge[['user_id', 'avg_price']].head(10)
```

```python
# create spending flag
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spend_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spend_flag']= 'Low spender'
```

```python
ords_prods_merge['spend_flag'].value_counts(dropna = False)
```

```python
# check accuracy of new column
ords_prods_merge[['user_id', 'avg_price', 'spend_flag']].head(10)
```

```python
# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 
# create new column
ords_prods_merge['order_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output
ords_prods_merge[['user_id', 'order_freq']].head(10)
```

```python
# create the frequency flag
ords_prods_merge.loc[ords_prods_merge['order_freq'] > 20, 'freq_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_freq'] > 10) & (ords_prods_merge['order_freq'] <= 20), 'freq_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_freq'] <= 10, 'freq_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['freq_flag'].value_counts(dropna = False)
```
